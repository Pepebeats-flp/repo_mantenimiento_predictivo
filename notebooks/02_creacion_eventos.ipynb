# 02. Creación de eventos

Agrupa correctivos ejecutados por bus y día para obtener eventos técnicos únicos y calcula el tiempo transcurrido desde el correctivo anterior.


```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / "src").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

DATA_RAW_DIR = PROJECT_ROOT / "data" / "raw"
DATA_PROCESSED_DIR = PROJECT_ROOT / "data" / "processed"
MODELS_DIR = PROJECT_ROOT / "models"
PLOTS_DIR = PROJECT_ROOT / "outputs" / "plots"
METRICS_DIR = PROJECT_ROOT / "outputs" / "metrics"

for directory in [DATA_PROCESSED_DIR, MODELS_DIR, PLOTS_DIR, METRICS_DIR]:
    directory.mkdir(parents=True, exist_ok=True)
```

```python
import pandas as pd
from IPython.display import display

from src.preprocessing import create_eventos_dataframe
```

```python
base_df = pd.read_parquet(DATA_PROCESSED_DIR / "base.parquet")
eventos_df = create_eventos_dataframe(base_df)
eventos_path = DATA_PROCESSED_DIR / "eventos.parquet"
eventos_df.to_parquet(eventos_path, index=False)

correctivos_base = base_df.loc[base_df["tipo_revision"].eq("CORRECTIVO")].copy()
intervalos = eventos_df["dias_desde_correctivo_anterior"].dropna()

print("Correctivos ejecutados en base:", len(correctivos_base))
print("Eventos técnicos únicos:", len(eventos_df))
print("Promedio días entre correctivos:", intervalos.mean())
print("Mediana días entre correctivos:", intervalos.median())
print("Parquet guardado en:", eventos_path)

display(eventos_df.head())
```
